```python
import pandas as pd

url = "https://raw.githubusercontent.com/wri/global-power-plant-database/master/output_database/global_power_plant_database.csv"
# We'll load the first 2000 rows to keep it snappy
df_power = pd.read_csv(url, nrows=2000)

# 1. See what kind of columns we have (Names, Fuel, Capacity)
print(df_power[['name', 'primary_fuel', 'capacity_mw']].head())

# 2. Demo: Group by fuel type and sum up the total Megawatts
fuel_summary = df_power.groupby('primary_fuel')['capacity_mw'].sum()

# 3. Visual: A Pie Chart of where the energy comes from
fuel_summary.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title="Energy Capacity by Fuel Type")
```
