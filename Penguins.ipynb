```python
import pandas as pd
from pyodide.http import open_url

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = open_url(url)
df = pd.read_csv(df)

print(df.describe())
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Create a boxplot: The line in the middle is the Median, 
# the edges of the box are the quartiles, and thea 'whiskers' show the Min/Max range.
sns.boxplot(data=df, x='species', y='bill_length_mm', palette="Set2")

# Add a title
plt.title("Distribution of Bill Length by Species (Showing Min, Max, and Median)")
plt.show()
```
